```python
#importing matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
 #plotting a basic graph(sine graph)
x = np.linspace(0,10,100)
#print(x)
y = np.sin(x)
#print(y)
plt.plot(x,y)
plt.title('sine graph')
plt.xlabel('values')
plt.ylabel('sine values')
plt.show()
```

```python
#cosine graph
import matplotlib.pyplot as plt
x = np.linspace(0,20,50)
y = np.cos(x)
plt.plot(x,y)
plt.title('cosine graph')
plt.xlabel('values')
plt.ylabel('cosine values')
plt.show()
```

PARABOLA


```python
x = np.linspace(-10,10,20)
y = x ** 2 #this is an equation for parabola
plt.plot(x,y)
plt.show()
```

PLOTTING WITH DIFFERENT SYMBOLS

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,'g+') # here g indicates colour and plus indicates the symbol we want to present
plt.show()
```

MULTIPLE GRAPHS

```python
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x),'g*')
plt.plot(x,np.cos(x),'r+')
plt.show()
''' or we can also write in this way
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,'g*')
plt.plot(x,z,'r+')
plt.show()'''
```

BAR GRAPH

```python
fig = plt.figure()
a = fig.add_axes([0,0,1,1]) # rectangele (here we can use any kind of coordinates)
languages = ['English','French','Spanish','Latin']
people = [100,200,300,400]
a.bar(languages,people)
plt.xlabel('languages')
plt.ylabel('people')
plt.title('Bar graph')
plt.show()
```

PIE CHART

```python
fig = plt.figure()
a = fig.add_subplot(111) # or fig.add_subplot(1,1,1)
languages = ['English','French','Spanish','Latin']
people = [100,200,300,400]
a.pie(people,labels = languages,autopct = '%1.1f%%')
plt.show()
```

SCATTER PLOT


```python
x = np.linspace(0,10,10)
y = np.sin(x)
z = np.cos(x)
plt.scatter(x,y)
plt.scatter(x,z)
plt.show()
```

3D SCATTER PLOT

```python
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
a = fig.add_subplot(111, projection='3d')  # it helps to print 3d figure
z = 20 * np.random.random(100) #it creates sprial form
x = np.sin(z)
y = np.cos(z)

a.scatter(x, y, z, c=z, cmap='summer')
plt.show()
```
